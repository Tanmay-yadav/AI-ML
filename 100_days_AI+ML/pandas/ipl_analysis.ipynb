# THe pandas plot() 


## what is bar chart 
## A bar chart is a graphical representation of data that uses rectangular bars to compare different categories or groups. Each bar's length or height corresponds to the value or frequency of the category it represents. Bar charts are commonly used to visualize categorical data, making it easy to see differences and trends among various groups. They can be displayed vertically or horizontally and are useful for comparing discrete data points across different categories.

```python
# visualization 
# plotting the number of matches won by each team
data['winner'].value_counts().plot(kind='bar')
```

```python
# horizontal bar plot
data['winner'].value_counts().plot(kind='barh')
# it will return all the matches played in Bangalore
```

```python
# bar chart for top 5 teams winning the matches
data['winner'].value_counts().head().plot(kind='bar')
```

```python
# pie chart for top 5 teams winning the matches
data['winner'].value_counts().plot(kind='pie')
# increase the size of pie chart
data['winner'].value_counts().plot(kind='pie',figsize=(10,10))
```

# The sort_values()

```python
(data['team1'].value_counts() +data['team2'].value_counts()).sort_values(ascending=False)
# sort_values(ascending=False) is used to sort the combined counts of matches played by each team in descending order.
```

```python
data.sort_values('city',ascending=False)
# inplace =true is used to permanently modify the original DataFrame without creating a new one.
# we have to be extra careful while using inplace =true as it will change the original data permanently.
```

```python
# how to create new data using inplace =true without creating new variable?
```

```python
data.sort_values(['city','date'],ascending=[True,False])
# it will sort the data first by 'city' in ascending order and then by 'date' in descending order within each city.
```

# The drop_duplicates() Method  


```python
data.drop_duplicates(subset=['city']).shape
# it will return the number of unique cities where matches were played
```

```python
data.drop_duplicates(subset=['city','season'])
# it will return the unique matches played between different teams on different dates
```

```python
# droping duplicates based on seasons and keep last 
data.drop_duplicates(subset=['season'],keep='last')[['season','winner']].sort_values('season')
```

# The Groupby() function 


```python
# THe Groupby function 
# the groupby function is used to group the data based on one or more columns and perform aggregate functions on the grouped data.
data.groupby('winner')['win_by_runs'].mean().sort_values(ascending=False)
data.head()
```

```python
data['win_by_runs'].sort_values(ascending=False).head()
# it will return all the matches win by runs
```

```python
# group the team1 and team2 columns to get total matches played by each team using groupby()
data.groupby('team1').size() + data.groupby('team2').size()
```

```python
# what is the use of size() function?
# size() function is used to get the number of elements in each group of a grouped DataFrame.   

```
