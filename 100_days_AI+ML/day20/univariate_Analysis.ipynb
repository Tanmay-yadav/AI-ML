# Univariate Analysis 

 * univariate analysis is the independent analysis of every column 

## Types of data 

* 1. Numerical data 
* 2. Categorical data

```python
import pandas as pd 
df =pd.read_csv('titanic.csv')
df.head()
# df.shape
```

## 1. Categorical Data

### a.Countplot 

```python
%pip install seaborn matplotlib
```

```python
df['Survived'].value_counts().plot(kind='bar') # this is used to plot chart of kind bar and we can 
# also do this with the help of seborn shown below 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
sns.countplot(x='Embarked',data =df) # countplot is used for categorical columns to see the count of each category   
plt.show()
```

### B. Piechart 


```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%2f')
```

## 2. Numerical Data

### a. Histogram

```python
plt.hist(df['Age'],bins=50) 
# The bins parameter in plt.hist() determines how many bars the data is divided into.
# Think of bins as "containers." If you are looking at the 'Age' column, 
# you are essentially telling Python: "Take the entire range of ages 
# (e.g., from 0 to 100) and split them into 50 equal intervals."
```

### b. Displot 

```python
sns.distplot(df['Age'],bins =50)
```

### c. Box plot


```python
sns.boxplot(df['Age'])
```

```python
df['Age'].min()
```

```python
df['Age'].max()
```

```python
df['Age'].mean()
```

```python
df['Age'].skew()
```
