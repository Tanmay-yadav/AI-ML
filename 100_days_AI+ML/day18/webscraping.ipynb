# Web Scraping

```python
# what is webscraping ?
# Web scraping is the automated process of extracting data from websites.   
# It involves using software tools or scripts to access web pages, parse the HTML content, and retrieve specific information for further analysis or use.
```

```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
# requests is used to make HTTP requests to fetch web pages.
# BeautifulSoup is a library used to parse HTML and XML documents, making it easier to navigate
# and extract data from web pages.
# beautifulsoup is used to parse the HTML content of the fetched web pages.
```

```python
headers ={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
import lxml 
print(lxml.__version__)
```

```python
soup=BeautifulSoup(webpage,"lxml")
# soup=BeautifulSoup(webpage,'html.parser')
soup.prettify()
# lxml is a parser that BeautifulSoup can use to parse HTML and XML documents.
# It is known for its speed and efficiency compared to the built-in 'html.parser'.
# Using lxml can lead to faster parsing times, especially for large documents.      
# To use lxml with BeautifulSoup, you need to have the lxml library installed in your Python environment.
# You can install it using pip:
# pip install lxml
# Once lxml is installed, you can specify it as the parser when creating a BeautifulSoup object:
# soup = BeautifulSoup(webpage, 'lxml')
# This tells BeautifulSoup to use the lxml parser to parse the HTML content of the webpage.
print(soup.title)
print(soup.title.text)
# print(soup.find_all('a'))  # to find all anchor tags
print(soup.find_all('a')[0])  # to find first anchor tag
```

```python
final = pd.DataFrame()
all_frames = [] # Use a list to store dataframes temporarily
for page in range(1,11):
    headers ={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}      
    webpage=requests.get(f'https://www.ambitionbox.com/list-of-companies?page={page}',headers=headers).text
    soup=BeautifulSoup(webpage,"lxml")
    company = soup.find_all('div', class_='companyCardWrapper')
    name =[]
    rating =[]
    review =[]
    services =[]
    location=[]
    Highly_rated_for=[]
    Critically_rated_for=[]
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div', class_='rating_text').text.strip())
        review.append(i.find("a", class_="companyCardWrapper__ActionWrapper").text)
        services.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[0])
        location.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[-1])
        # Highly_rated_for.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('\n')[0])
        # Critically_rated_for.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip())
    # len(name)
    # print(rating)
    # print(name)
    # print(review)
    # print(services)
    # print(location)
    # print(Highly_rated_for)
    # print(Critically_rated_for)

    # Create DF for the CURRENT page
    page_df = pd.DataFrame({
        'name': name, 
        'rating': rating, 
        'review': review, 
        'services': services, 
        'location': location
    })
    
    # Append the DF to our list
    all_frames.append(page_df) 


# FINALLY, combine everything once the loop is done
final = pd.concat(all_frames, ignore_index=True)
    
```
