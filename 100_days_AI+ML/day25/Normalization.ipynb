```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # plotting
import seaborn as sns # statistical data visualization
from sklearn.metrics import accuracy_score  
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('E:\\WORK\\AI-ML\\data\\wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class','Alcohol','Malic_acid']
```

```python
df['Class'].value_counts().plot(kind='bar')
plt.show()
```

```python
sns.kdeplot(df['Alcohol'],shade=False)
```

```python

sns.kdeplot(df['Malic_acid'],shade=False)
```

```python
color_dict={1:'red',2:'blue',3:'green'}
sns.scatterplot(
    x='Alcohol',
    y='Malic_acid',
    hue='Class',
    data=df,
    palette=color_dict
)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Class',axis=1),df['Class'],test_size=0.3,random_state=0)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler() 
# fit the scaler to the train set,it will learn the parameters 
scaler.fit(X_train)
# transform both train and test set
X_train_scaled=scaler.transform(X_train)    
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(10,5))
ax1.scatter(X_train['Alcohol'],X_train['Malic_acid'],c=y_train)
ax1.set_title('Before Normalization')
ax2.scatter(X_train_scaled['Alcohol'],X_train_scaled['Malic_acid'],c=y_train)
ax2.set_title('After Normalization')
plt.show()  
```

```python
fig,(ax1,ax2)= plt.subplots(ncols=2,figsize=(12,5))
# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alcohol'],shade=True,ax=ax1,label='Alcohol')
sns.kdeplot(X_train['Malic_acid'],shade=True,ax=ax1,label='Malic_acid')
# after scaling 
ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['Alcohol'],shade=True,ax=ax2,label='Alcohol')
sns.kdeplot(X_train_scaled['Malic_acid'],shade=True,ax=ax2,label='Malice_acid')
plt.show()
```
